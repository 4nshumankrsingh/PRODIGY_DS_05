```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = r'C:\Users\Anshuman\Downloads\Delhi Accident Data.csv'
accident_data = pd.read_csv(file_path, dtype=str)
```

```python
irrelevant_columns = ['Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9']
for col in irrelevant_columns:
    if col in accident_data.columns:
        accident_data = accident_data.drop(columns=[col])

accident_data_cleaned = accident_data.dropna()
```

```python
yearly_accidents = accident_data_cleaned['YEAR'].value_counts().sort_index()
district_accidents = accident_data_cleaned['DISTRICT'].value_counts()
type_of_accidents = accident_data_cleaned['TYPE OF ACCIDENT'].value_counts()
vehicles_at_fault = accident_data_cleaned['VEHICLE AT FAULT'].value_counts()
injured = accident_data_cleaned['# INJURED'].astype(int).sum()
killed = accident_data_cleaned['# KILLED '].astype(int).sum()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=yearly_accidents.index, y=yearly_accidents.values)
plt.title('Number of Accidents per Year')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=district_accidents.index, y=district_accidents.values)
plt.title('Number of Accidents by District')
plt.xlabel('District')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=type_of_accidents.index, y=type_of_accidents.values)
plt.title('Type of Accidents')
plt.xlabel('Type of Accident')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=vehicles_at_fault.index, y=vehicles_at_fault.values)
plt.title('Vehicles at Fault')
plt.xlabel('Vehicle at Fault')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```

```python
print(f"Total number of injured: {injured}")
print(f"Total number of killed: {killed}")
```
